##  Binary Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
def SigmoidBinary(t):
    return 1/(1+np.exp(-t))
t = np.linspace(-10, 10)
plt.plot(t, SigmoidBinary(t))
plt.title('Binary Sigmoid Activation Function')
plt.show()
```

# # leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
def LeakyRectifiedLinearUnit(arr):
    lst=[]
    for i in arr:
        if i>=0:
            lst.append(i)
        else:
            lst.append(0.01*i)
    return lst
arr = np.linspace(-5, 5)
plt.plot(arr, LeakyRectifiedLinearUnit(arr))
plt.title('Leaky ReLU Activation Funtion')
plt.show()
```

# # TanH

```python
import numpy as np 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
def TanH(t):
    return 2/(1+np.exp(-t)-1)
t = np.linspace(-5,5)
plt.plot(t, TanH(t))
plt.title('TanH Activation Function')
plt.show()
```
